6 . What is the difference between a Run object and a Paragraph object?

A: A document contains multiple paragraphs. A paragraph begins on a new line and contains multiple runs. Runs are contiguous groups of characters within the a paragraph.

9 . What is the difference between False, True, and None for the bold variable?

A: True (the attribute is always enabled, no matter what other styles are applied to the run), 

False (the attribute is always disabled), 

or None (defaults to whatever the run’s style is set to).

10 . How do you create a Document object for a new Word document?

11 . How do you add a paragraph with the text &#39;Hello, there!&#39; to a Document object stored in a
variable named doc?

```python
import docx
doc = docx.Document()
doc.add_paragraph('Hello, there!')
doc.save('helloworld.docx')
```
